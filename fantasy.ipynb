```python
## Import necessary packages
import numpy as np
import scipy as sp
import pandas as pd
```

```python
## Read in excel data for 2016 player stat projections
ff = pd.read_excel("fantasyprojections_082316.xlsx")
```

```python
## Show a sample of the dataframe
ff[ff.rec_att > 0].head(5)
```

```python
## Fantasy for Real Boys scoring system
ff['points'] = (ff.pass_yds*0.04 + ff.pass_tds*4 - ff.pass_ints*2 \
                + ff.rush_yds*0.1 + ff.rush_tds*6 + ff.rec_yds*0.1 \
                + ff.rec_tds*6 + ff.rec_att*1 - ff.fumbles*2) 
```

```python
## Sort the dataframe by points
ff_sorted = ff.sort_values(['points'], ascending=False)
```

```python
## Average number of points for each position
ff_sorted.groupby('pos').mean().points
```

```python
## Average number of points for the top 24 at each position
ff_sorted.groupby('pos').head(24).groupby('pos').mean().points
```

```python
## Get the mean starter points
mean_starter_qb = ff_sorted[ff_sorted.pos == 'QB'].head(12).mean().points
mean_starter_rb = ff_sorted[ff_sorted.pos == 'RB'].head(24).mean().points
mean_starter_wr = ff_sorted[ff_sorted.pos == 'WR'].head(24).mean().points
mean_starter_flex = ff_sorted[ff_sorted.pos.isin(('WR','RB'))].head(60).mean().points
mean_starter_te = ff_sorted[ff_sorted.pos == 'TE'].head(12).mean().points
print "QB: %f" % mean_starter_qb
print "RB: %f" % mean_starter_rb
print "WR: %f" % mean_starter_wr
print "Flex: %f" % mean_starter_flex
print "TE: %f" % mean_starter_te
```

```python
## Set the mean value for each starter by position
ff_sorted.loc[ff_sorted.pos == 'QB','mean_starter_points'] = mean_starter_qb
ff_sorted.loc[ff_sorted.pos == 'RB','mean_starter_points'] = mean_starter_rb
ff_sorted.loc[ff_sorted.pos == 'WR','mean_starter_points'] = mean_starter_wr
ff_sorted.loc[ff_sorted.pos.isin(('WR','RB')),'mean_flex_points'] = mean_starter_flex
ff_sorted.loc[ff_sorted.pos == 'TE','mean_starter_points'] = mean_starter_te
```

```python
## Create point_bonus column showing how many points above average a player scores
ff_sorted["point_bonus"] = ff_sorted.points - ff_sorted.mean_starter_points
```

```python
## Sort by point_bonus
top_players = ff_sorted.sort_values('point_bonus',ascending=False)
```
